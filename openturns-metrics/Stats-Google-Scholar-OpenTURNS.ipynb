# Google Scholar references for OpenTURNS

In Google Scholar, we searched for "openturns uncertainties" from 2007 to 2019.

## References

https://scholar.google.fr/scholar?q=openturns+uncertainties&hl=fr&as_sdt=1%2C5&as_vis=1&as_ylo=2019&as_yhi=2019

```python
import numpy as np
import openturns as ot
import openturns.viewer as otv
```

```python
data = [[2007, 5],
        [2008, 3],
        [2009, 14],
        [2010, 15],
        [2011, 20],
        [2012, 20],
        [2013, 31],
        [2014, 33],
        [2015, 35],
        [2016, 39],
        [2017, 40],
        [2018, 57],
        [2019, 58]
       ]
```

```python
data = np.array(data)
```

```python
sample = ot.Sample(data)
sample.setDescription(["Year", "Google Scholar"])
```

```python
graph = ot.Graph("References of \"openturns uncertainties\" on Google Scholar",
                 "Year", "Annual number", True)
cloud = ot.Cloud(sample)
graph.add(cloud)
graph
```

```python
view = otv.View(graph)
view.save("openturns-googlescholar-messages.png")
```
