# UQ Software Google Scholar metrics

We want to compare the rise of UQ software as mentioned in research papers. To do this, we searched in Google Scholar. 
- The word "OpenTURNS" is confusing because it is the name used to designate a swimming technique. This is why we searched for "OpenTURNS uncertainty". 
- The name "URANIE" can be associated with several other topics, including the name of a ship that was used to go to Nouvelle Calédonie in the 1800s. The word "URANIE uncertainties" is not specific enough ([example](https://scholar.google.fr/scholar?q=uranie+uncertainties&hl=fr&as_sdt=1%2C5&as_vis=1&as_ylo=2000&as_yhi=2000)). This is why we used "URANIE Gaudier" ([example](https://scholar.google.fr/scholar?hl=fr&as_sdt=1%2C5&as_ylo=2012&as_yhi=2012&as_vis=1&q=URANIE+gaudier&btnG=)) because Fabrice Gaudier is one of the core developer of the library: citing the URANIE software means citing one of the reference paper that F.Gaudier wrote. 
- The name "Dakota" is a US state, so that we cannot simply search for this particular name. We searched for "Dakota Eldred" because M.S. Eldred is one of the authors of several papers on this software.

In Google Scholar, we searched for the following keywords starting from 2007:
- "OpenTURNS uncertainties" ([Example](https://scholar.google.fr/scholar?hl=fr&as_sdt=1%2C5&as_ylo=2012&as_yhi=2012&as_vis=1&q=URANIE+gaudier&btnG=));
- "URANIE Gaudier"; ([Example](https://scholar.google.fr/scholar?q=URANIE+gaudier&hl=fr&as_sdt=1%2C5&as_vis=1&as_ylo=2009&as_yhi=2009))
- "Dakota Eldred optimization" [Example](https://scholar.google.fr/scholar?hl=fr&as_sdt=1%2C5&as_ylo=2005&as_yhi=2005&as_vis=1&q=dakota+Eldred+optimization&btnG=)
- "UQLab Sudret" [Example](https://scholar.google.fr/scholar?q=uqlab+sudret&hl=fr&as_sdt=1%2C5&as_vis=1&as_ylo=2015&as_yhi=2015)

```python
import numpy as np
import openturns as ot
import openturns.viewer as otv
```

```python
sampleOpenTURNS = ot.Sample.ImportFromCSVFile("OpenTURNS-Google-Scholar.csv")
sampleURANIE = ot.Sample.ImportFromCSVFile("URANIE-Google-Scholar.csv")
sampleDakota = ot.Sample.ImportFromCSVFile("Dakota-Google-Scholar.csv")
sampleUQLab = ot.Sample.ImportFromCSVFile("UQLab-Google-Scholar.csv")
```

```python
def plotSample(data, legend, pointStyle):
    cloud = ot.Cloud(data)
    cloud.setPointStyle(pointStyle)
    cloud.setLegend(legend)
    return cloud
```

```python
graph = ot.Graph("References of UQ softwares on Google Scholar",
                 "Year", "Annual number", True)
graph.add(plotSample(sampleOpenTURNS, "OpenTURNS", "circle"))
graph.add(plotSample(sampleURANIE, "URANIE", "diamond"))
graph.add(plotSample(sampleDakota, "DAKOTA", "triangleup"))
graph.add(plotSample(sampleUQLab, "UQLab", "triangledown"))
graph.setColors(ot.Drawable_BuildDefaultPalette(len(graph.getDrawables())))
graph.setLegendPosition("topleft")

view = otv.View(graph)
```
