```python
import openturns as ot
from IPython.display import display, HTML
import SuperProjectionStrategy

ot.Log.Show(ot.Log.NONE)
```

```python
ot.RandomGenerator.SetSeed(0)
dimension = 2
input_names = ["x1", "x2"]
formulas = ["cos(x1 + x2)", "(x2 + 1) * exp(x1 - 2 * x2)"]
model = ot.SymbolicFunction(input_names, formulas)
distribution = ot.Normal(dimension)
x = distribution.getSample(3000)
y = model(x)

algo = ot.FunctionalChaosAlgorithm(x, y, distribution)
algo.run()
```

```python
projectionStrategy = algo.getProjectionStrategy()
projectionStrategy
```

```python
projectionImplementation = projectionStrategy.getImplementation()
print(type(projectionImplementation))
projectionImplementation
```
